```python
import pandas as pd
import numpy as np 
import warnings 
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv(r'https://raw.githubusercontent.com/krishnaik06/Types-Of-Cross-Validation/main/cancer_dataset.csv')
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.drop('Unnamed: 32',axis=1,inplace=True)
```

```python
data.isnull().sum()
```

```python
x = data.iloc[:,2:]
y = data.iloc[:,1]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold, cross_val_score

k = KFold(5)
model = LogisticRegression()
```

```python
cross_val_score(model,x,y,cv=k)
```

```python
log = LogisticRegression()
arr = [2,3,4,5,6]
for i in arr:
    k =  KFold(i)
    result = cross_val_score(log,x,y,cv = k)
    print('when k = {} , acc = {} '.format(i,np.mean(result)))
```

```python
#leave one out method
```

```python
from sklearn.model_selection import LeaveOneOut
lcv = LeaveOneOut()
result = cross_val_score(model,x,y,cv = lcv)
```

```python
np.mean(result)
```

# Hyper parameter tuning

```python
from sklearn import svm, datasets
iris = datasets.load_iris()
df = pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df
```

```python
df['target'] = iris.target
```

```python
df
```

```python
kernels = ['rbf','linear']
c = [1,10,20,30]
avg_scores = {}
for kval in kernels:
    for cval in c:
        cv_score = cross_val_score(svm.SVC(kernel =kval,C=cval,gamma='auto'),iris.data,iris.target,cv = 5)
        avg_scores[kval + '_'+ str(cval)] = np.mean(cv_score)
```

```python
(avg_scores)
```

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(svm.SVC(gamma = 'auto'),{'C':[1,10,20,30],'kernel' :['rbf','linear','sigmoid','poly'],'max_iter':[1,5,8,10,20]},cv=10)
clf.fit(iris.data,iris.target)
                   
```

```python
DF = pd.DataFrame(clf.cv_results_)
DF
```

```python
clf.best_score_
```

```python
clf.best_params_
```

```python
from sklearn.model_selection import RandomizedSearchCV
r_cv = RandomizedSearchCV(svm.SVC(gamma='auto'),{'C':[1,10,20,30],'kernel' :['rbf','linear','sigmoid','poly'],'max_iter':[1,5,8,10,20]},cv=10,random_state=5)
r_cv.fit(iris.data,iris.target)
```

```python
pd.DataFrame(r_cv.cv_results_)
```

```python
r_cv.best_estimator_
```
