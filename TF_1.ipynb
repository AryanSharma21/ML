```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

#  RANDOM DATA GENERATION 

```python
observations = 1000
xs = np.random.uniform(low=-10, high = 10, size=(observations,1))
xz = np.random.uniform(-10, 10, size=(observations,1))

generated_inputs = np.column_stack((xs,xz))
noise = np.random.uniform(-1,1,(observations,1))

generated_targets = 2*xs + 3*xz +5 + noise

np.savez( 'TF_intro', inputs = generated_inputs, targets=generated_targets)
```

# LOAD THE DATA

```python
training_data = np.load(r'C:\Users\aryan\Desktop\datasets\TF_intro.npz')
```

```python
training_data
```

```python
input_size = 2
output_size = 1

model = tf.keras.Sequential([
                            tf.keras.layers.Dense(output_size)
                            ])

model.compile(optimizer='sgd', loss='mean_squared_error')

model.fit(training_data['inputs'], training_data['targets'] , epochs=200, verbose=2)
```
