# FEATURE SEPARATION

```python
x_features = data.iloc[:,1:3]
x_features
```

#  IMPORTING KMEANS

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(4)
```

```python
found_cluster = kmeans.fit_predict(x_features)
```

```python
data_with_cluster = data.copy()
data_with_cluster['cluster'] = found_cluster
data_with_cluster
```

```python
data_with_cluster.tail(10)
```

#  PLOTTING CLUSTER

```python
plt.scatter(data_with_cluster["Longitude"],data_with_cluster['Latitude'], c = data_with_cluster['cluster'],cmap='rainbow')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
